```python
import sys
import os

sys.path.append(os.path.abspath("../../src"))

from data.load_data import load_data

chocolate_df = load_data("../../data/processed/chocolate_price.csv")

dark_chocolates = chocolate_df[chocolate_df["Cocoa Percent"] > 70].reset_index()

has_Trinitario = dark_chocolates["Bean Type"].str.contains("Trinitario", case=False, na=False)

for idx, val in enumerate(has_Trinitario):
    if val == True:
        ten_percent = dark_chocolates.iloc[idx]["price(100g)"] / 10
        dark_chocolates.at[idx, "price(100g)"] += ten_percent

priceSum = dark_chocolates["price(100g)"].sum()


dark_chocolates.to_csv("../../data/processed/dark_chocolates.csv")

dark_chocolates
```
